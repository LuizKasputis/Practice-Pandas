```python
#2 e 3
usuarios = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep ='|', index_col = 'user_id')
usuarios.head(10)
```

```python
#4
ocupacao = usuarios.groupby('occupation').mean()
ocupacao
```

```python
#5 Esse exercicio é interessante
def sexo(genero):
    if(genero == 'M'):
        return 1
    if(genero == 'F'):
        return 0
    
#Crio um data frame novo, que deixa entre 0 e 1 se é masculino ou feminino
usuarios['novo_genero'] = usuarios['gender'].apply(sexo)

porcentagem = usuarios.groupby('occupation').novo_genero.sum()/usuarios.occupation.value_counts()*100

porcentagem.sort_values(ascending = False)
```
