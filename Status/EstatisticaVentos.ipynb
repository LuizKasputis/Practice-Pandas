```python
#2 e 3 Usando date time, eu consigo dizer quais colunas fazem partes da estrutura datetime
data = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/Wind_Stats/wind.data', sep = '\s+', parse_dates = [[0,1,2]] )
data
```

```python
#4 criar uma função que que ajusta o ano
def ajuste(x):
    if(x.year > 1989):
        year = x.year - 100
    else:
        year = x.year
    return datetime.date(year, x.month, x.day)

data['Yr_Mo_Dy'] = data['Yr_Mo_Dy'].apply(ajuste)

data.head(10)
```

```python
#5
#Não entendi pq ele fez esse trecho, eu joguei direto e deu certo. Mas não sei se mais pra frente da problema
data["Yr_Mo_Dy"] = pd.to_datetime(data["Yr_Mo_Dy"])

data = data.set_index('Yr_Mo_Dy')

data.head()
```
