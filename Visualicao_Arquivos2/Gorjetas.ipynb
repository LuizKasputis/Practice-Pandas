```python
#1
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

%matplotlib inline
```

```python
# 2 e 3
tips = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv', sep = ',')
tips
```

```python
#4
del tips['Unnamed: 0']
```

```python
tips
```

```python
#5 jeito q eu fiz
df = tips.total_bill.sort_values(ascending = False)

gorjeta = np.arange(0,55,5)

plt.hist(df, bins = gorjeta)

plt.xlabel('Valor Pago')
plt.ylabel('Frequencia')
plt.title('Valor pago Frequencia')

plt.show()
```

```python
#Do gabarito
ttbill = sns.distplot(tips.total_bill);
ttbill.set(xlabel = 'Valor', ylabel = 'Frequencia', title = "Valor total")
sns.despine()
```

```python
#6
scatter = sns.lmplot(x = 'total_bill', y = 'tip', data = tips, fit_reg = False)
scatter.set(title = 'Valor da conta X Gorjeta')
```

```python
#7
sns.pairplot(tips)
```

```python
#8
sns.stripplot(x = "day", y = "total_bill", data = tips);
```

```python
#9
lm = sns.stripplot(x = 'tip', y = 'day', data = tips, hue = 'sex')
lm.set(title = 'Gorjeta Paga X Sexo')
```

```python
#10
sns.boxplot(x = "day", y = "total_bill", hue = "time", data = tips);
```

```python
#11
sns.set(style = "ticks")
g = sns.FacetGrid(tips, col = "time")
g.map(plt.hist, "tip");
```

```python
#12
g = sns.FacetGrid(tips, col = "sex", hue = "smoker")
g.map(plt.scatter, "total_bill", "tip", alpha =.7)

g.add_legend();

```
