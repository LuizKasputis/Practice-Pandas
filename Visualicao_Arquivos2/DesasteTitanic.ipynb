```python
#1
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline
```

```python
#2 e 3
titanic = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Titanic_Desaster/train.csv', sep = ',')
titanic.head()
```

```python
#4
titanic.set_index('PassengerId')
```

```python
#5
masculino = titanic[titanic.Sex == 'male']
feminino = titanic[titanic.Sex == 'female']

lista = [len(masculino), len(feminino)]

df = pd.DataFrame({'Sexo': ['Male', 'Female'], 'Porcentagem': lista, })

df.set_index('Sexo')
```

```python
plt.pie(
    df.Porcentagem,
    labels = df.Sexo,
    shadow = False,
    colors = ['blue','red'],
    explode = (0.15 , 0),
    startangle = 90,
    autopct = '%1.1f%%'
)
plt.axis('equal')

plt.title("Sex Proportion")

plt.tight_layout()
plt.show()
```

```python
#6 Para obter identificar o genero tbm é preciso utilizar o lmplot

lm = sns.lmplot(x = 'Age', y = 'Fare', data = titanic, hue = 'Sex', fit_reg = False)
lm.set(title = 'Idade x Tarifa Paga')

axes = lm.axes
axes[0,0].set_ylim(-5,)
axes[0,0].set_xlim(-5,85)
```

```python
#8 Pra fazer esse tem que criar um vetor que separa os valores, e cria uma margem, apartir disso tem que jogar no eixo X.

df = titanic.Fare.sort_values(ascending = False)


binsVal = np.arange(0,600,10)

plt.hist(df, bins = binsVal)

plt.xlabel('Tarifa')
plt.ylabel('Frequencia')
plt.title('Frequncia por tarifa')

plt.show()
```
